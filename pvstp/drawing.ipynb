# Tutorial
Сначала рассчитаем, сколько итераций требуется для сходимости сети.
Для этого есть фукнция:
find_convergence(количество node, количество edge, max_simul_steps, vlans_n, verbose=True, rstp=True)
Функция вовзращает:
- экземпляр класса Network, в котором хранится топология графа,
чтобы потом его можно было использовать чтобы нарисовать сеть.
- cls - количество шагов, необходимых для сходимости для конкретного vlan
- dim - диаметр графа у конкретного vlan


```python

# Постоим простую сеть с 1 vlan на 3 узла и 3 дуги
vlans_n = 1
nets, c1s, d1s = find_convergence(3, 3, 100, vlans_n, rstp=False, verbose=True)
```

Нарисуем процесс, для этого реализована draw_stp(nets, название гифки, конец симуляции, количество влан)

```python
# Строится столько графиков, сколько vlans. Каждый график сверху назван соотв.
# номером vlan. Синим помечены вершины, которые имеют выход на хосты с данным
# влан, черным - те, которые не имеют.
draw_stp(nets, 'test-3', 15, vlans_n)
```

Сломаем какой-то линк

```python
# Если broke_times не равен None, то заускается симуляция до того момента, когда не наступает
# момент broke_times + 1 (логично его задать как convergence+1)
# И рвется линк по номером edgeid
# Затем симуляция длиться до max_step
# Сломанный линк помечается черным цветом
max_step = 40

draw_stp(nets, 'test-cut-3', max_step, vlans_n, broke_times=c1s, edgeid=2)
```
