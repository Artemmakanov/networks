# Tutorial
Сначала рассчитаем, сколько итераций требуется для сходимости сети.
Для этого есть фукнция:
find_convergence(количество node, количество edge, max_simul_steps, verbose=True, rstp=True)
Функция вовзращает:
- экземпляр класса Network, в котором хранится топология графа,
чтобы потом его можно было использовать чтобы нарисовать сеть.
- convergence - количество шагов, необходимых для сходимости
- dim - диаметр графа


```python
# построим сеть с 3 вершинами, 3 линками, максимальное время до схождения алгоритма - 100
# verbose=True - включить tqdm, rstp=False - используем сеттинг ванильного stp
net, convergence, dim = find_convergence(3,3, 100, verbose=True, rstp=False)
print(f"Stp сошелся на {convergence} шаге, диаметр дерева = {dim}")
```

draw_stp(net, имя, convergence+1) - принимает в качестве агрументов net, название для гифки,
и момент, когда следует прекратить рисование (время схождения + 1)

```python
# нарисуем процесс, произощедший в ячейке выше
# FORWARD_DELAY (время между состояния портов), установлен как 5 итераций.
draw_stp(net, 'test-3', convergence+1)
# Вершины помечаются как 
# bid
# root bid
# rpc
```

draw_stp(net, имя, кол-во шагов симуляции, broke_time=время, когда режется линк, edgeid=номер линка)

```python
# Если broke_time не равен None, то заускается симуляция до того момента, когда не наступает
# момент broke_time (логично его задать как convergence+1)
# И рвется линк по номером edgeid (можно также задать линк через bid1 и bid2).
# Затем симуляция длиться до max_step
# Сломанный линк помечается черным цветом и крестиком
max_step = 40
draw_stp(net, 'test-cut-3', max_step, broke_time=convergence+1, edgeid=1)
```

```python
# также можно активировать лог:
# построим сеть с 3 вершинами, 3 линками, максимальное время до схождения алгоритма - 100
# verbose=True - включить tqdm, rstp=False - используем сеттинг ванильного stp
net, convergence, dim = find_convergence(3,3, 100, verbose=True, rstp=False, log=True)
print(f"Stp сошелся на {convergence} шаге, диаметр дерева = {dim}")
```
